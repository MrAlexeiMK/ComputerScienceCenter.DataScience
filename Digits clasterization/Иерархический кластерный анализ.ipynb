```python
import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
matplotlib.style.use('ggplot')
%matplotlib inline
```

```python
data = pd.read_csv("digit.dat", sep=";")

#Удаляем лишние столбцы
data.pop("A")
data.pop("A2")
data.pop("B2")
data.pop("C2")
data.pop("D2")
data.pop("E2")
data.pop("F2")
data.pop("G2")
data.pop("H2")

#Заменяем строки на числа
data.replace("ZERO", 0, inplace=True)
data.replace("zero", 0, inplace=True)
data.replace("ONE", 1, inplace=True)
data.replace("one", 1, inplace=True)
data.replace("two", 2, inplace=True)
data.replace("three", 3, inplace=True)
data.replace("four", 4, inplace=True)
data.replace("five", 5, inplace=True)
data.replace("six", 6, inplace=True)
data.replace("seven", 7, inplace=True)
data.replace("eight", 8, inplace=True)
data.replace("nine", 9, inplace=True)

data.head()
```

```python
link = linkage(data, 'ward', 'euclidean')
```

```python
fig = plt.figure(figsize=(12,12))

#Строим и выводим дендограмму
dn = dendrogram(link, color_threshold = -1)
```

```python
#Поэкспериментируем с числом кластеров

#10 кластеров
data['cluster'] = fcluster(link, 10, criterion='maxclust')
data.groupby('cluster').mean()

#Охарактеризуем каждый кластер
#Кластер 1 - это цифра 4 (преобладают C, D, E, G. Эти буквы характеризуют линии, описывающие цифру 4, с остальными аналогично)
#Кластер 2 - это какой-то выброс, скорее всего туда вошли ошибки в данных
#Кластер 3 - это цифра 1
#Кластер 4 - это цифра 5
#Кластер 5 - это цифра 7
#Кластер 6 - это цифра 6
#Кластер 7 - это цифра 2
#Кластер 8 - это цифра 8
#Кластер 9 - это цифра 9
#Кластер 10 - это цифра 3
#Не хватает цифры 0
```

```python
#7 кластеров
data['cluster'] = fcluster(link, 7, criterion='maxclust')
data.groupby('cluster').mean()

#Охарактеризуем каждый кластер
#1 кластер - цифра 4
#2 кластер - цифры 1, 7
#3 кластер - неясно как интерпретировать
#4 кластер - цифры 5, 6
#5 кластер - цифра 2
#6 кластер - цифры 0, 8
#7 кластер - цифры 3, 9
#Непонятно, как интерпретировать 3-ий кластер
```

```python
#6 кластеров
data['cluster'] = fcluster(link, 6, criterion='maxclust')
data.groupby('cluster').mean()

#Охарактеризуем каждый кластер
#1 кластер - цифра 4
#2 кластер - цифры 1, 7
#3 кластер - цифры 5, 6
#4 кластер - цифра 2
#5 кластер - цифры 0, 8
#6 кластер - цифры 3, 9
#Похожие цифры попати в одинаковые кластеры
```

```python
#5 кластеров
data['cluster'] = fcluster(link, 5, criterion='maxclust')
data.groupby('cluster').mean()

#Охарактеризуем каждый кластер
#1 кластер - цифра 4
#2 кластер - цифры 1, 7
#3 кластер - цифры 5, 6
#4 кластер - цифра 2
#5 кластер - цифры 0, 8, 3, 9
#Слишком много цифр вошло в последний кластер
```

```python
#4 кластера
data['cluster'] = fcluster(link, 4, criterion='maxclust')
data.groupby('cluster').mean()

#Охарактеризуем каждый кластер
#1 кластер - цифра 1, 4, 7
#2 кластер - цифры 5, 6
#3 кластер - цифра 2
#4 кластер - цифра 0, 8, 3, 9
```

Поэксперементировав с числом кластером я решил остановиться на 6-ти кластерах, т.к. цифры в них разбросаны наиболее равномерно и все кластеры хорошо интерпретируются

```python
#Строим график каменистой осыпи
dist = link[:, 2]
idxs = range(1, len(dist) + 1)
plt.plot(idxs, dist, marker='o')
plt.title('Расстояние между объединяемыми кластерами')
plt.xlabel('Шаг')
plt.ylabel('Расстояние')
plt.axhline(8.5, c='k', linestyle='dashed')
plt.text(25, 9, '6 кластеров')
```
