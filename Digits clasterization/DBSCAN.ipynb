```python
import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from sklearn.cluster import DBSCAN
matplotlib.style.use('ggplot')
%matplotlib inline
```

```python
data = pd.read_csv("digit.dat", sep=";")

#Удаляем лишние столбцы
data.pop("A")
data.pop("A2")
data.pop("B2")
data.pop("C2")
data.pop("D2")
data.pop("E2")
data.pop("F2")
data.pop("G2")
data.pop("H2")

#Заменяем строки на числа
data.replace("ZERO", 0, inplace=True)
data.replace("zero", 0, inplace=True)
data.replace("ONE", 1, inplace=True)
data.replace("one", 1, inplace=True)
data.replace("two", 2, inplace=True)
data.replace("three", 3, inplace=True)
data.replace("four", 4, inplace=True)
data.replace("five", 5, inplace=True)
data.replace("six", 6, inplace=True)
data.replace("seven", 7, inplace=True)
data.replace("eight", 8, inplace=True)
data.replace("nine", 9, inplace=True)

data.head()
```

```python
dbscan = DBSCAN(eps=0.5, metric='euclidean', min_samples=5)
dbscan.fit(data)
```

```python
unique, counts = np.unique(dbscan.labels_, return_counts=True)
print(np.asarray((unique, counts)).T)
3#Слишком много кластеров и выбрасов
```

```python
#Переберём min_samples
for i in range(1, 30):
    for j in np.arange(0.01, 2, 0.01):
        dbscan = DBSCAN(eps=j, metric='euclidean', min_samples=i)
        dbscan.fit(data)

        unique, counts = np.unique(dbscan.labels_, return_counts=True)
        if len(np.asarray((unique, counts)).T) > 3 and len(np.asarray((unique, counts)).T) < 8: print(np.asarray((unique, counts)).T, i, j)
#Основываясь на предыдущих квастерных анализах попробуем взять случай с min_samples=26, где 6 кластеров
```

```python
#Зафиксируем min_samples 26 и поперебираем eps
for i in np.arange(0.01, 2, 0.01):
    dbscan = DBSCAN(eps=i, metric='euclidean', min_samples=26)
    dbscan.fit(data)

    unique, counts = np.unique(dbscan.labels_, return_counts=True)
    print(np.asarray((unique, counts)).T, i)
#Заметим, что при eps > 1.41 резко меняется число кластеров на 3, а при остальных случаях вообще ничего не меняется, оставим eps=0.5
```

```python
#Проинтерпретируем
dbscan = DBSCAN(eps=0.5, metric='euclidean', min_samples=26)
dbscan.fit(data)

unique, counts = np.unique(dbscan.labels_, return_counts=True)
print(np.asarray((unique, counts)).T, i)

data['dbscan'] = dbscan.labels_
data.groupby('dbscan').mean()
#Кластер 0 - цифра 
#Кластер 1 - цифра 5
#Кластер 2 - цифра 9
#Кластер 3 - цифра 8
#Кластер 4 - цифра 6
#Кластер 5 - цифра 3
```
